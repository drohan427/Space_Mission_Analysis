 # ***Data Preparation***

 ***Identify missing values and handle it?***

```python
df.isnull().sum()
```

```python
df[' Rocket']=df[' Rocket'].fillna(0)
df
```

 ***Rename Columns if needed?***

```python
df=df.rename(columns={" Rocket":"Price"})
df
```

```python
df=df.rename(columns={"Detail":"Rocket"})
df
```

 ***Convert datatypes if needed?***

```python
df.Price=df.Price.replace('[,]','',regex=True).astype("float64")
```

***Extract month and year column?***

```python
df['Year']=df.Datum.str[12:16]
df['Month']=df.Datum.str[4:8]
df
```

# ***Top Launching Agencies:***

***Which organizations have conducted the most launches?***

```python
cn=df.CompanyName.value_counts().nlargest(10)
```

```python
plt.figure(figsize=(12,6))
sb.barplot(x=cn.index, y=cn.values, color='skyblue')  
plt.xlabel('Organisation', fontsize=12)
plt.ylabel('Number of Launches', fontsize=12)
plt.title('Top Companies by Launch Count', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

***How many successfull space missions has each company conducted?***

```python
df[df.StatusMission == 'Success'].groupby('CompanyName').size().sort_values(ascending=False).nlargest(10)
```

***Who has the highest number of failed or partially failed missions?***

```python
df[df.StatusMission == 'Failure'].groupby('CompanyName').size().sort_values(ascending=False).nlargest(10)
```

# ***Launch Site Analysis:***

# ***Temporal Trends:***

***How has the number of space launches changed over the Years?***

```python
yr=df.Year.value_counts().sort_index()
```

```python
plt.figure(figsize=(10,6))
sb.lineplot(x=yr.index, y=yr.values, color='skyblue')
plt.xticks(rotation=90)
plt.show()
```

***Are there any noticeable trends in mission outcomes over the years?***

```python
a2=df.groupby(['Year','StatusMission']).size().reset_index()
a2
```

```python
plt.figure(figsize=(18,8))
sb.barplot(x=a2.Year,y=a2[0],hue=a2.StatusMission,palette='pastel')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of Missions', fontsize=14)
plt.title('Missions per Year by Status', fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.legend(title='Mission Status', bbox_to_anchor=(1.01, 1.01))
plt.show()
```
